```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfs = []
for file in os.listdir("data/downloads/drive"):
    xlsx = pd.read_excel(f"data/downloads/drive/{file}", parse_dates=["AppointmentDate"])
    dfs.append(xlsx)
data = pd.concat(dfs)
```
